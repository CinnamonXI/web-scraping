```python
import pandas as pd
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen, Request
```

```python
base_url = "https://veritasihomes.com/premium-homes/"
extend = ["camberwall-advantage/", "camberwall-advantage-3/", "camberwall-advantage-4/"]
url = base_url+extend[0]
req = Request(url, headers = {"User-Agent": 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.15; rv:78.0) Gecko/20100101 Firefox/78.0'})
resp = urlopen(req)
html = resp.read()
```

```python
soup = bs(html)
print(soup.prettify())
```

```python
location = soup.find("span", class_="elementor-icon-list-text").string
location
```

```python
name = soup.find("h2", class_= "elementor-heading-title").string
name
```

```python
description = soup.find("div", class_="elementor-column-wrap").string
description
```
